# Carga de datos a través de la función open

```python
data3 = open(mainpath + "\\" + "customer-churn-model\\Customer Churn Model.txt", "r")
```

```python
cols = data3.readline().strip().split(",")
n_cols = len(cols)
```

```python
counter = 0
main_dict =  {}
for col in cols:
    main_dict[col] = []
```

```python
for line in data3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1

print("El dataset tiene %d filas y %d columnas"%(counter, n_cols))
```

```python
df3= pd.DataFrame(main_dict)
df3.head()
```

```python
infile = mainpath + "\\" + "customer-churn-model\\Customer Churn Model.txt"
outfile = mainpath + "\\" + "customer-churn-model\\Tab Customer Churn Model.txt"
```

```python
#Separar los diferentes elementos mediate un tabulador (/t)
with open(infile, "r") as infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```

```python
df4 = pd.read_csv(outfile, sep="\t")
df4.head()
```


# Leer datos desde una URL

```python
medals_url = "http://winterolympicsmedals.com/medals.csv"
```

```python
medals_data = pd.read_csv(medals_url)
```

```python
medals_data.head()
```

```python
import csv
import urllib3
```

```python
response = urllib3.PoolManager().request('GET', medals_url)
response_csv = csv.reader(response.data.decode('utf-8').split('\n'))
```

```python
medals_cols = next(response_csv)
medals_n_cols = len(medals_cols)
medals_cols
```

```python
count_med = 0
medal_dict = {}
for col in medals_cols:
    medal_dict[col] = []
```

```python
for line in response_csv:
    for i in range(medals_n_cols):
        medal_dict[medals_cols[i]].append(line[i])
    count_med += 1
print("El data set tiene %d filas y %d columnas"%(count_med, medals_n_cols))
```
